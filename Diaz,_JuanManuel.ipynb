```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np


from google.colab import files

# Permitir al usuario cargar un archivo
archivo_cargado = files.upload()

# Imprimir los nombres de los archivos cargados
for nombre_archivo in archivo_cargado.keys():
   print(f"Archivo cargado: {nombre_archivo}")



#  ruta del archivo Excel
avion_archivo = "Airplane_Crashes_and_Fatalities_Since_1908.xlsx"

# Abrir el archivo Excel
df_avion = pd.read_excel(avion_archivo, engine='openpyxl')

# Muestramos los registros
df_avion

```

**Haremos un data cleaning**

```python
#Limpiamos la data
datalimpia = df_avion.fillna(0)
datalimpia
```

Mathplotlib

**Crear un grafico que nos muestre la evolucion de los accidentes a traves de los años.**

```python
# datos por año y contamos la cantidad de accidentes por año
datalimpia['Date'] = pd.to_datetime(datalimpia['Date'])
accidentes_por_año = datalimpia.groupby(datalimpia['Date'].dt.year).size()

# Creamos el gráfico de líneas
plt.plot(accidentes_por_año.index, accidentes_por_año.values)

# Personalizamos el gráfico
plt.title('Cantidad de Accidentes de Avión por Año (1908-2017)')
plt.xlabel('Años de Accidentes')
plt.ylabel('Cantidad de Accidentes por Año')

# Mostramos el gráfico
plt.show()
```

**Se realizara un grafico de barras sobre los accidentes, separandolo en  periodos de 10 años**

```python




# Convertimos la columna 'date' a tipo de datos datetime
datalimpia['Date'] = pd.to_datetime(df_avion['Date'])

# Filtramos los datos para que solo incluyan años entre 1908 y 2017
df_filtrado = datalimpia[(datalimpia['Date'].dt.year >= 1908) & (datalimpia['Date'].dt.year <= 2017)]

# Agrupamos los datos en períodos de 10 años y calculamos la suma de la cantidad de accidentes en cada período
accidentes_por_periodo = df_filtrado.groupby(pd.Grouper(key='Date', freq='10Y')).size()

# Creamos el gráfico de barras
accidentes_por_periodo.plot(kind='bar')

# Personalización del gráfico
plt.title('Cantidad de Accidentes de Avión por Período de 10 Años (1908-2017)')
plt.xlabel('Período de 10 Años')
plt.ylabel('Cantidad de Accidentes')

# Modificar las etiquetas del eje X para mostrar los períodos de 10 años
periodos = [(str(año_inicio) + '-' + str(año_inicio + 9)) for año_inicio in range(1908, 2009, 10)]
plt.xticks(range(len(periodos)), periodos, rotation=45)

# Mostrar el gráfico
plt.show()

```

**Se realizara un grafico que muestre la diferencia entre sobrevientes y victimas en accidentes de aviones entre 1908-2017**


```python

# Calculamos la cantidad de sobrevivientes
datalimpia['Sobrevivientes'] = datalimpia['Aboard'] - datalimpia['Fatalities']


# Crear el gráfico de líneas
plt.plot(datalimpia['Date'], datalimpia['Fatalities'], label='Víctimas')
plt.plot(datalimpia['Date'], datalimpia['Sobrevivientes'], label='Sobrevivientes')

# Personalización del gráfico
plt.title('Accidentes de Avión (1908-2017)')
plt.xlabel('Años')
plt.ylabel('Cantidad')
plt.legend()

# Mostrar el gráfico
plt.show()
```

**Victimas de Accidentes de Avion por Decada**

```python

# Convertimos la columna 'Date' a tipo de datos datetime
datalimpia['Date'] = pd.to_datetime(datalimpia['Date'])

# Agrupamos los datos en períodos de 10 años y calcula la cantidad de víctimas en cada período
datalimpia['Decade'] = datalimpia['Date'].dt.year // 10 * 10
victimas_por_decada = datalimpia.groupby('Decade')['Fatalities'].sum().reset_index()

# Creamos el gráfico
plt.figure(figsize=(10, 6))
sns.barplot(x='Decade', y='Fatalities', data=victimas_por_decada, hue='Decade', dodge=False, palette='viridis', legend=False)

# Personalización del gráfico
plt.title('Víctimas de Accidentes de Avión por Década (1908-2017)')
plt.xlabel('Década')
plt.ylabel('Cantidad de Víctimas')

# Mostrar el gráfico
plt.show()

```

**Sobrevivientes de Accidentes de Avion por Decada**

```python


# Convertimos la columna 'Date' a tipo de datos datetime
datalimpia['Date'] = pd.to_datetime(datalimpia['Date'])

# Calculamos el numero de Sobrevientes
datalimpia['Survivors'] = datalimpia['Aboard'] - datalimpia['Fatalities']

# Agrupamos los datos en períodos de 10 años y calcula la cantidad de víctimas en cada período
datalimpia['Decade'] = datalimpia['Date'].dt.year // 10 * 10
survivors_per_decade = datalimpia.groupby('Decade')['Survivors'].sum().reset_index()

# Creamos el gráfico
plt.figure(figsize=(10, 6))
sns.barplot(x='Decade', y='Survivors', data=survivors_per_decade, hue='Decade', dodge=False, palette='viridis', legend=False)

# Personalización del gráfico
plt.title('Sobrevientes de Accidentes de Avion por Decada (1908-2017)')
plt.xlabel('Decada')
plt.ylabel('Numero de Sobrevivientes')

#Mostramos el Grafico
plt.show()


```
